```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Load the dataset
fuel_econ = pd.read_csv('data/fuel_econ.csv')

# Display first few rows of the dataset
fuel_econ.head(10)

# Create a histogram
plt.figure(figsize=(10, 6))
plt.hist(fuel_econ['co2'], bins=30, edgecolor='black')
plt.title('Distribution of Carbon Dioxide Emissions')
plt.xlabel('Carbon Dioxide Emissions (g/km)')
plt.ylabel('Frequency')
plt.savefig('plots/histogram.png')
plt.show()


# Create a heatmap of correlations
numeric_data = fuel_econ.select_dtypes(include=[np.number])
correlation_matrix = numeric_data.corr()

# The heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=False, fmt=".2f", cmap='RdBu', square=True, cbar_kws={"shrink": .8})
plt.title('Heatmap of Correlation Matrix')


plt.savefig('correlation_matrix_heatmap.png')
plt.show()



# Interpretation of diagrams
# Add comments or markdown cells explaining the insights from the plots
```

The Interpretations

The distribution of the CO2 emission

General Shape: The histogram shows a unimodal distribution with a peak between approximately 350 and 450 g/km of CO2 emissions. This suggests that most vehicles in the dataset emit CO2 in this range.

Frequency Distribution: There is a gradual rise in the number of vehicles emitting between 200 and 350 g/km, indicating these emission levels are relatively common. The frequency decreases sharply beyond 450 g/km, indicating fewer vehicles with very high CO2 emissions.

Implications:The data suggests that vehicle emissions are clustered around mid-range values (350-450 g/km). This can indicate that the majority of vehicles are within a typical efficiency range, with fewer vehicles being extremely fuel-efficient or extremely inefficient.

The heatmap correlation matrix

Fuel Efficiency Metrics (city, highway, comb, etc.) are highly positively correlated with each other, meaning vehicles efficient in one metric tend to be efficient in others. Engine Size (displ, cylinders) is strongly negatively correlated with fuel efficiency and positively correlated with CO2 emissions, indicating larger engines consume more fuel and emit more CO2. CO2 Emissions (co2) are inversely correlated with fuel efficiency, showing that more efficient vehicles produce less CO2. Greenhouse Gas Score (ghgScore) and Fuel Economy Score (feScore) are positively correlated with fuel efficiency and negatively with CO2 emissions.
